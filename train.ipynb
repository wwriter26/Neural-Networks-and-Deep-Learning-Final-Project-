## TRAINING FILE:
- this file is going to load the Kaggle ASL dataset via ImageFolder, shuffle the 80/10/10 split of training, validation, testing, applies the augmentations that we mentioned (flips, brightness +-20%, rotation +-10˚), trains the 4-block CNN (going from 32 -> 64 -> 128 -> 256 channels) with Adam/CrossEntropy for 15 EPOCHS. It will save the Best_model.pth on best val accuracy and prints teh top 1 and top5 test accuracy, and generates a Confusion matrix

#### 1. Dataset and splits

- Load the **Kaggle ASL** dataset using PyTorch **`ImageFolder`** (directory layout: class name → images).
- Build an **80% / 10% / 10%** split for **training**, **validation**, and **testing**, with data **shuffled** appropriately before or during split assignment.

#### 2. Augmentations

Training transforms (as discussed):

- **Flips** (horizontal/vertical as you choose and document).
- **Brightness** jitter roughly **±20%**.
- **Rotation** up to about **±10°**.

#### 3. Model architecture

A **four-block CNN** with channel widths **32 → 64 → 128 → 256** (each block stacks conv layers and downsampling as you define). Classification head matches the number of ASL classes.

#### 4. Training setup

- **Optimizer:** Adam  
- **Loss:** cross-entropy (multi-class classification)  
- **Epochs:** **15**  
- Train on the training split; monitor **validation** accuracy each epoch for model selection.

#### 5. Saving the best model

Persist weights to **`Best_model.pth`** (or your chosen path) whenever **validation accuracy** improves over the best so far—reload this checkpoint for final test evaluation.

#### 6. Evaluation

On the **held-out test** set:

- Report **top-1** and **top-5** accuracy.  
- Plot a **confusion matrix** (predicted vs. true class) for error analysis.
